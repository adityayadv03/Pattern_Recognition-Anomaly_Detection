```python
#Numpy Operation
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
data=iris.data
feature_names=iris.feature_names
```

```python
#Basic Operation on Arrays
arr1 = np.arange(4, dtype = np.float_).reshape(2, 2)
print('First array:')
print(arr1)
print('\nSecond array:')
arr2 = np.array([12, 12])
print(arr2)
print('\nAdding the two arrays:')
print(np.add(arr1, arr2))
print('\nSubtracting the two arrays:')
print(np.subtract(arr1, arr2))
print('\nMultiplying the two arrays:')
print(np.multiply(arr1, arr2))
print('\nDividing the two arrays:')
print(np.divide(arr1, arr2))
```

```python
#Mean Calculation
mean=np.mean(data,axis=0)
print("Mean of Each feature is :- ")
for name, mean in zip(feature_names, mean):
    print(f"{name}: {mean:.3f}")
```

```python
#Median Calculation
median=np.median(data,axis=0)
print("Median of Each feature is :- ")
for name, median in zip(feature_names, median):
    print(f"{name}: {median:.3f}")
```

```python
#Standard Deviation
std=np.std(data,axis=0)
print("Standard Deviation of Each feature is :- ")
for name, std in zip(feature_names, std):
    print(f"{name}: {std:.3f}")
```

```python
#Sum
sum=np.sum(data,axis=0)
print("Sum of Each feature is :- ")
for name, sum in zip(feature_names, sum):
    print(f"{name}: {sum:.3f}")
```

```python
#Highest Total Measurement
total_measurement=np.sum(data,axis=0)
max_index=np.argmax(total_measurement)
print(f"\nSample with the Highest Total Measurement:")
print(f"Index: {max_index}, Total Measurement: {total_measurement[max_index]:.1f}")
```

```python
#Sorting Data by Sepal Length
sorted_data = data[np.argsort(data[:, 0])]  # Sort by the first column (Sepal Length)
print("\nSorted Dataset by Sepal Length (First Few Rows):")
print(sorted_data[:10])
```

```python
#Variance Calculation
variance=np.var(data,axis=0)
print("Variance of Each feature is :- ")
for name , var in zip(feature_names,variance):
  print(f"{name}: {var:.3f}")
```

```python
#Pandas Operation
import pandas as pd
#Basic Operation
data = {'Name':['Tom', 'nick', 'krish', 'jack'],
        'Age':[20, 21, 19, 18]}
df=pd.DataFrame(data)
print(df)
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.describe()
```

```python
#Sum Most Used Operation
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, np.nan, 45, 56],
        'Third Score':[52, 40, 80, 98],
        'Fourth Score':[np.nan, np.nan, np.nan, 65]}
dff=pd.DataFrame(dict)
print(dff.isnull())
print(dff.fillna(dff['First Score'].mean()))
df[25:50]
```

```python
#Grouping the data and Calculating the Mean
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
print("Mean of Each Feature Grouped by Species:")
print(df.groupby('species').mean())
```

```python
#Counting the Number of Sample for Each Species
print("Count of Samples for Each Species:")
print(df['species'].value_counts())
```

```python
#Adding a New Column for Total Measurement
df['total_measurement'] = df.iloc[:, :-1].sum(axis=1)
print("DataFrame with Total Measurement Column :")
print(df.head())
```

```python
#Sorting the Dataset by sepal Width
sorted_df=df.sort_values(by='sepal width (cm)', ascending=False)
print("Dataset Sorted by Sepal Length :")
print(sorted_df.head())
```

```python
#Filtering Row where petal length greater than 1.5
filtered_df = df[df['petal length (cm)'] > 1.5]
print("Filtered Rows Where Petal Width > 1.5:")
print(filtered_df)
```
